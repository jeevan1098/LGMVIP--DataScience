```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv(r'/content/globalterrorismdb_0718dist.csv',encoding="latin")
data.head()
```

```python
data.describe()
```

```python
columns = data.columns.to_list()
columns
```

```python
data1 = pd.DataFrame(data, data.rename(columns={ 'success': 'Success', 'location': 'Location', 'iyear':'Year','imonth':'Month','extended':'Extended','iday':'Day','country_txt':'Country',
                   'country': 'Country_code', 'provstate':'State','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target',
                   'nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                   'weaptype1_txt':'Weapon_type','motive':'Motive','city':'City','latitude':'Latitude','longitude':'Longitude'},inplace=True))
```

```python
data1.info()
```

```python
data1.head(),data1.tail()
```

```python
data1.isnull().sum()
```

```python
df=data1['City'].value_counts().to_frame().drop('Unknown').sort_values('City',axis=0,ascending=False).head(10)
df.plot(kind='bar',figsize=(20,10),color='darkgreen')
plt.xticks(rotation = 50)
plt.xlabel("City",fontsize=15)
plt.ylabel("Total attacks",fontsize=15)
plt.title("Top 10 most affected cities",fontsize=20)
plt.show()
```

```python
plt.subplots(figsize=(20,10))
index1=data1['Country'].value_counts()[:10].index
values1=data1['Country'].value_counts()[:10].values
sns.barplot(index1,values1,palette='rocket')
plt.title('Top 10 most affected countries', fontsize = 20)
plt.xlabel('Country', fontsize = 15)
plt.ylabel('Total attacks', fontsize = 15)
plt.xticks(rotation = 50)
plt.show()
```

```python
df1=data1['Region'].value_counts().to_frame().sort_values('Region',axis=0,ascending=False).head(10)
df1.plot(kind='bar',figsize=(20,10),color='blue')
plt.xticks(rotation = 50)
plt.xlabel("Region",fontsize=15)
plt.ylabel("Total attacks",fontsize=15)
plt.title("Top 10 most affected regions",fontsize=20)
plt.show()
```

```python
df2=data1[['Group','Killed']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('Killed',ascending=False).head(10)
df2.plot(kind='bar',color='grey',figsize=(20,10))
plt.title("Top 10 terrorist attack groups",fontsize=20)
plt.xlabel("Terrorist group",fontsize=20)
plt.ylabel("Total attacks",fontsize=20)
plt.xticks(rotation=50)
plt.show()
```

```python
data1['AttackType'].value_counts().plot(kind='bar',figsize=(20,10),color='pink')
plt.xticks(rotation = 50)
plt.xlabel("Attack Type",fontsize=15)
plt.ylabel("Total attacks",fontsize=15)
plt.title("Most used attack types",fontsize=20)
plt.show()
```
